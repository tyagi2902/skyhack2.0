```python
import pandas as pd
```

```python
calls_df = pd.read_csv('callsclean.csv')
customers_df = pd.read_csv('customers.csv')
reason_df = pd.read_csv('reason.csv')
sentiment_df = pd.read_csv('sentiment_statistics.csv')
```

```python
calls_df['call_start_datetime'] = pd.to_datetime(calls_df['call_start_datetime'], format='%d-%m-%Y %H:%M', errors='coerce')
calls_df['call_end_datetime'] = pd.to_datetime(calls_df['call_end_datetime'], format='%d-%m-%Y %H:%M', errors='coerce')
calls_df['agent_assigned_datetime'] = pd.to_datetime(calls_df['agent_assigned_datetime'], format='%d-%m-%Y %H:%M', errors='coerce')
```

```python
# 1. Calculate AHT
# Convert call_start_datetime and call_end_datetime to datetime
calls_df['call_start_datetime'] = pd.to_datetime(calls_df['call_start_datetime'])

```

```python
calls_df['call_end_datetime'] = pd.to_datetime(calls_df['call_end_datetime'])
```

```python
# Calculate call_duration in seconds
calls_df['call_duration'] = (calls_df['call_end_datetime'] - calls_df['agent_assigned_datetime']).dt.total_seconds()
```

```python
# 2. Merge DataFrames to include reasons and sentiments
merged_df = calls_df.merge(reason_df, on='call_id', how='inner') \
                     .merge(sentiment_df, on='call_id', how='inner')
```

```python
# 3. Calculate AHT by primary_call_reason
aht_by_reason = merged_df.groupby('primary_call_reason')['call_duration'].mean().reset_index()
aht_by_reason.columns = ['primary_call_reason', 'average_aht']
```

```python
# 4. Identify most and least frequent call reasons
frequency_by_reason = merged_df['primary_call_reason'].value_counts().reset_index()
frequency_by_reason.columns = ['primary_call_reason', 'frequency']
```

```python
# Merge to get both frequency and AHT
frequency_aht_df = frequency_by_reason.merge(aht_by_reason, on='primary_call_reason')
```

```python
# Identify most and least frequent call reasons
most_frequent_reason = frequency_aht_df.loc[frequency_aht_df['frequency'].idxmax(), 'primary_call_reason']
least_frequent_reason = frequency_aht_df.loc[frequency_aht_df['frequency'].idxmin(), 'primary_call_reason']
```

```python
# Calculate AHT for most and least frequent call reasons
most_frequent_aht = aht_by_reason.loc[aht_by_reason['primary_call_reason'] == most_frequent_reason, 'average_aht'].values[0]
least_frequent_aht = aht_by_reason.loc[aht_by_reason['primary_call_reason'] == least_frequent_reason, 'average_aht'].values[0]
```

```python
aht_by_reason
```

```python
# 5. Calculate percentage difference in AHT
percentage_difference = ((most_frequent_aht-least_frequent_aht) / most_frequent_aht) * 100
```

```python
# Print results
print(f"Most Frequent Call Reason: '{most_frequent_reason}' with AHT: {most_frequent_aht:.2f} seconds")
print(f"Least Frequent Call Reason: '{least_frequent_reason}' with AHT: {least_frequent_aht:.2f} seconds")
print(f"Percentage difference in AHT: {percentage_difference:.2f}%")
```
